### Pertanyaan 1: Bagaimana performa penjualan dan revenue perusahaan selama ini?

```python
df['month_year'] = df['order_approved_at'].dt.to_period('M')

sales_performance = df.groupby('month_year').agg({
    'order_id': 'count',         # Number of orders
    'payment_value': 'sum'       # Total revenue
}).reset_index()

sales_performance['month_year'] = sales_performance['month_year'].dt.to_timestamp()

plt.figure(figsize=(12, 8))

blue = "#72BCD4"

plt.subplot(2, 1, 1)
sns.lineplot(x='month_year', y='order_id', data=sales_performance, marker='o', color=blue)
plt.title('Trend of Number of Orders per Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.xticks(sales_performance['month_year'], rotation=45, ha='right')  # Display all months on the X-axis

# Plot for total revenue
plt.subplot(2, 1, 2)
sns.lineplot(x='month_year', y='payment_value', data=sales_performance, marker='o', color=blue)
plt.title('Trend of Total Revenue per Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(sales_performance['month_year'], rotation=45, ha='right')  # Display all months on the X-axis

plt.tight_layout()
plt.show()
```

**Insight:**
Dari Oktober 2016 hingga sekitar November 2017, terlihat tren pertumbuhan positif dengan peningkatan pesanan yang signifikan. Lonjakan pesanan terjadi terutama pada November 2017. Setelah mencapai puncak pada bulan tersebut, jumlah pesanan menurun sedikit tetapi tetap stabil dengan fluktuasi kecil hingga pertengahan 2018. Tidak ada transaksi yang diproses pada November 2016. Pada September 2018, terjadi penurunan drastis dalam jumlah pesanan, yang mungkin disebabkan oleh faktor musiman atau perubahan signifikan dalam strategi atau kondisi pasar.

### Pertanyaan 2: Produk apa yang paling banyak dan paling sedikit terjual?

```python
# Menghitung jumlah produk yang terjual per kategori
sum_order_items_df = df.groupby("product_category_name_english").order_id.nunique().sort_values(ascending=False).reset_index()
sum_order_items_df.head(10)
```

```python
# Warna
blue = '#8EACCD'
green = '#D2E0FB'
colors = [blue, green, green, green, green]

# Menghitung jumlah produk yang terjual per kategori
product_sales = df.groupby('product_category_name_english')['order_id'].count().reset_index()
product_sales.columns = ['product_category_name_english', 'total_sold']
product_sales = product_sales.sort_values(by='total_sold', ascending=False)

# Membuat subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

# Produk dengan penjualan terbanyak
sns.barplot(x="total_sold", y="product_category_name_english", 
            data=product_sales.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product", loc="center", fontsize=18)
ax[0].tick_params(axis='y', labelsize=15)

# Menambahkan label nilai pada batang
for i in ax[0].patches:
    ax[0].text(i.get_width() + 0.2, i.get_y() + i.get_height() / 2, 
               f'{int(i.get_width())}', ha='right', va='center', fontsize=12, color='black', weight='bold')

# Produk dengan penjualan paling sedikit
sns.barplot(x="total_sold", y="product_category_name_english", 
            data=product_sales.sort_values(by="total_sold", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()  # Membalikkan sumbu x
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()  # Memindahkan label sumbu y ke kanan
ax[1].set_title("Worst Performing Product", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

# Menambahkan label nilai pada batang
for i in ax[1].patches:
    ax[1].text(i.get_width() - 0.2, i.get_y() + i.get_height() / 2, 
               f'{int(i.get_width())}', ha='left', va='center', fontsize=12, color='black', weight='bold')

# Menambahkan judul utama
plt.suptitle("Best and Worst Performing Product Category by Number of Sales", fontsize=20)
plt.tight_layout()
plt.show()
```

## Analisis RFM

- Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.
- 
Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi
- 
Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

```python
rfm_df = df.groupby(by="customer_id", as_index=False).agg({
    "order_approved_at": "max", # mengambil tanggal order terakhir
    "order_id": "nunique", # menghitung jumlah order
    "payment_value": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
rfm_df 
```

```python
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = df["order_approved_at"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
rfm_df.head()
```

```python
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
rfm_df.describe()
```

```python
rfm_df.sort_values(by="recency", ascending=True).head(5)
```

```python
rfm_df.sort_values(by="frequency", ascending=False).head(5)
```

```python
# Mengambil 5 karakter customer_id di awal
rfm_df['short_customer_id'] = rfm_df['customer_id'].apply(lambda x: x[:5])  

# Plotting with shortened customer_id
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#8EACCD", "#8EACCD", "#8EACCD", "#8EACCD", "#8EACCD"]

sns.barplot(y="recency", x="short_customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

sns.barplot(y="frequency", x="short_customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sns.barplot(y="monetary", x="short_customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

for axis in ax:
    axis.tick_params(axis='x', rotation=45)
    
plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.show()
```

Mengurutkan customer berdasarkan recency, frequency, & monetary score

```python
rfm_df['r_rank'] = rfm_df['recency'].rank(ascending=False)
rfm_df['f_rank'] = rfm_df['frequency'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['monetary'].rank(ascending=True)

rfm_df.head()
```

```python
rfm_df['r_rank_norm'] = (rfm_df['r_rank']/rfm_df['r_rank'].max())*100
rfm_df['f_rank_norm'] = (rfm_df['f_rank']/rfm_df['f_rank'].max())*100
rfm_df['m_rank_norm'] = (rfm_df['m_rank']/rfm_df['m_rank'].max())*100
 
rfm_df.drop(columns=['r_rank', 'f_rank', 'm_rank'], inplace=True)
 
rfm_df.head()
```

```python
rfm_df['RFM_score'] = 0.15*rfm_df['r_rank_norm']+0.28 * \
    rfm_df['f_rank_norm']+0.57*rfm_df['m_rank_norm']
rfm_df['RFM_score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['customer_id', 'RFM_score']].head(7)
```

Segmentasi customer berdasarkan RFM_score

```python
rfm_df["customer_segment"] = np.where(
    rfm_df['RFM_score'] > 4.5, "Top customers", (np.where(
        rfm_df['RFM_score'] > 4, "High value customer",(np.where(
            rfm_df['RFM_score'] > 3, "Medium value customer", np.where(
                rfm_df['RFM_score'] > 1.6, 'Low value customers', 'lost customers'))))))

rfm_df[['customer_id', 'RFM_score', 'customer_segment']].head(20)
```

```python
customer_segment_df = rfm_df.groupby(by="customer_segment", as_index=False).customer_id.nunique()
customer_segment_df
```

```python
customer_segment_df['customer_segment'] = pd.Categorical(customer_segment_df['customer_segment'], [
    "lost customers", "Low value customers", "Medium value customer",
    "High value customer", "Top customers"
])
```

```python
plt.figure(figsize=(24, 5))
blue = '#8EACCD'
green = '#D2E0FB'
colors = [green, blue, green, green, green]
sns.barplot(
    x="customer_id", 
    y="customer_segment",
    data=customer_segment_df.sort_values(by="customer_segment", ascending=False),
    palette=colors
)
plt.title("Number of Customer for Each Segment", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

**Insight:**
- Lost customers merujuk pada pelanggan yang sudah tidak aktif berbelanja lagi. Meskipun jumlahnya sedikit, masih ada kesempatan untuk menarik mereka kembali dengan merancang strategi promosi yang menarik, seperti menawarkan penawaran spesial atau mengingatkan mereka tentang produk terbaru.

- Low value customers merupakan segmen yang paling banyak jumlahnya. Mereka masih melakukan pembelian, tetapi tidak sering dan dengan nilai yang rendah. Segmen ini memiliki potensi besar; dengan menawarkan promosi menarik atau program loyalitas, perusahaan dapat mendorong mereka untuk berbelanja lebih sering atau dengan nilai yang lebih tinggi.

- Medium value customers adalah pelanggan yang sudah cukup sering berbelanja, dengan nilai transaksi yang cukup baik. Mereka berada di posisi menengah, dan jika diberikan perhatian lebih, seperti penawaran eksklusif atau layanan yang lebih baik, mereka bisa bertransformasi menjadi pelanggan bernilai tinggi.

- High value customers adalah pelanggan yang sering berbelanja dengan nilai transaksi yang besar. Mereka sangat penting bagi bisnis karena kontribusi mereka terhadap pendapatan perusahaan. Untuk mempertahankan kesetiaan mereka, perusahaan perlu memberikan pelayanan istimewa, seperti akses prioritas ke produk baru atau diskon khusus.

- Top customers adalah pelanggan terbaik yang memberikan kontribusi terbesar. Meskipun jumlahnya sedikit, mereka adalah pelanggan paling setia dan sering melakukan transaksi besar. Mereka harus mendapatkan perhatian khusus melalui layanan VIP, penawaran eksklusif, atau penghargaan agar tetap loyal.
